## 5.	Analyze the system defined by the input-output relation y[n]=g[n]x[n] for linearity property and time-invariance property, when
### a) g[n] = 1 for all n	 
### b) g[n]	= n for all n 

```julia
# First we create the different functions needed as per the equation

# x(n) is taken to represent a cosine wave 
function x(n)
    return cos(2*pi*n)
end
# xn(n) is taken to represent a sine wave
function xn(n)
    return sin(2*pi*n)
end
# y(n) is our input using cos
function y(n)
    return g(n)*x(n)
end
# y(n) is our input using sine
function yn(n)
    return g(n)*xn(n)
end
# The parameter in the given question

# We take user input and choose whether we want to return 1 or n (given in question)
print("Enter Your choice (1 or 2)")
a = readline()
val=parse(Int64,a)
function g(n)
  if val==1
     return 1
  elseif val==2 
     return n
  else
    print("Wrong option")
      
    end
end
```

```julia
# Checking Time invariance property
using Plots
using LinearAlgebra
t=1:0.1:10
# delay
d=0.5
# input delay
inpdel=zeros(length(t))
inpdel=[x(t[i]-d) for i in 1:length(t)];
```

```julia
# Output delay
outdel=zeros(length(t))
outdel=[y(t[i]-d) for i in 1:length(t)];
if inpdel==outdel
    print("The equation is time invariant")
else
    print("The equation is time variant")
end
```

```julia
# Plotting the input and output delay
plot(t,outdel)
scatter!(t,inpdel)
```

```julia
# To check Linearity

# we create 2 equations
eq1=zeros(length(t))
eq2=zeros(length(t))

# Using additive property we give values to the two equations and compare
eq1=[(g(t[i])*x(t[i]))+g(t[i])*xn(t[i]) for i in 1:length(t)];
eq2=[y(t[i])+yn(t[i]) for i in 1:length(t)];

if eq1==eq2
    print("The equation is linear")
else
    print("The equation is non-linear")
end
```
