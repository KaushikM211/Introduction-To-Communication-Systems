```julia
using LinearAlgebra
using DSP
using FFTW
using Plots
```

# Question-1

```julia
#Signal Duration
T = 0.5;
#Sampling Interval
dt = 0.001;
#Sampling Frequency
fs = 1/dt;
#Time Vector
t = 0:dt:T;
#Message/Source/Information Frequency
fi = 10;
#Carrier Frequency
fc = 250;
#Single-Tone Message Signal
TX_signal = [sin(2*pi*fi*t[i]) for i in 1:length(t)];
#Carrier Signal
carrier = [cos(2*pi*fc*t[i]) for i in 1:length(t)]; 
#The AM-DSB-SC Modulated Signal
AM_DSB_SC = TX_signal.*carrier;
# Spectrum 
#Fourier Transform of Message/Source/Information Signal
M = fft(TX_signal); 
M = [abs(fftshift(M)[i]) for i in 1:length(M)];
#Fourier Transform of Modulated Signal
U = fft(AM_DSB_SC);
U = [abs(fftshift(U)[i]) for i in 1:length(U)];
#Length of Modulated Signal
n = length(AM_DSB_SC);
#Frequency Range
f = (-n/2:n/2-1)*(fs/n); 
```

```julia
#Plotting in the input/source/message signal
plot(t, TX_signal,label="Message Signal")
title!("Message Signal")
xlabel!("Time")
ylabel!("Amplitude")
```

```julia
#Plotting the carrier signal
plot(t, carrier, label="Carrier Signal")
title!("Carrier Signal")
xlabel!("Time")
ylabel!("Amplitude")
```

```julia
#Plotting the AM-DSB-SC Modulated Signal
plot(t, AM_DSB_SC,label="AM-DSB-Modulated Signal")
title!("AM-DSB-SC Modulated Signal")
xlabel!("Time")
ylabel!("Amplitude")
```

```julia
#Message+Modulated Signal
plot(t, TX_signal, label="Message Signal")
plot!(t, AM_DSB_SC, label="AM-DSB-SC Modulated Signal")
title!("Message+Modulated Signal")
xlabel!("Time")
ylabel!("Amplitude")
```

```julia
#Spectrum plot for Message/Information/Source Signal
plot(f,M,label="Spectrum")
title!("Message Signal Spectrum")
xlabel!("Frequency")
```

```julia
#Spectrum plot for AM-DSB-SC Modulated Signal
plot(f,U,label="Spectrum")
title!("AM-DSB-SC Modulated Signal Spectrum")
xlabel!("Frequency")
```
