# Q1)Generate and display elementary signals

```julia
#Unitstep function
using Plots
using LinearAlgebra
t = -5:0.1:5;
u=zeros(length(t));
for i=1:length(t)
    if t[i]>0
        u[i]=1
    else
        u[i]=0
    end
end
u
plot(t,u)
```

```julia
#Rectangular function
using Plots
using LinearAlgebra
t = -1:0.01:1;
x=zeros(length(t));
i1 = findall(t -> t == -0.5,t);
i2 = findall(t -> t == 0.5,t);
for i = i1[1]:i2[1]
x[i] = 1;
end
plot(t,x)
```

```julia
# Signum function
using Plots
using LinearAlgebra
t = -5:0.01:5;
x=zeros(length(t));
for i=1:length(t)
    if t[i]>0
        x[i]=1
    elseif t[i]<0
        x[i]=-1
    else
        x[i]=0
    end
end
x
plot(t,x)
```

```julia
# Exponential Signal
using Plots
using LinearAlgebra
t= -1:0.1:2;
y1=zeros(length(t));
y2=zeros(length(t));
a = 1;
for i in 1:length(t)
    y1[i]=exp(a*t[i]);
    y2[i]=exp(-a*t[i]);
end
plot(t,y1)
```

# Q2)Generate and display a periodic square waveform of frequency 5Hz.

```julia
using Plots
st_pt = 0;
end_pt = 2*pi;
delt = 0.01;
t = st_pt:delt:end_pt;
length(t);
freq = 5;
T = 2*pi/freq;
cycles = floor(end_pt/T)
indx_1 = convert(Int64,round(length(t)/(2*cycles),digits = 0)); 
T = 2*pi/freq;
f_t = zeros(length(t));
for i = 1:2:2*freq
    f_t[(i*indx_1)+1:((i+1)*indx_1)-1].= 1.0;
end
plot(t,f_t)
```

# Q3)Generate and display a rectangular aperiodic pulse

```julia
using Plots
st_pt = 0;
end_pt = 20;
delt = 0.01;
t = st_pt:delt:end_pt;
#Intialsing array
f_t = zeros(length(t));
#Finding index at t = 3 & 8
ind1 = findall(t -> t == 3,t);
ind2 = findall(t -> t == 8,t);
for i = ind1[1]:ind2[1]
f_t[i] = 1;
end
#plotting rectangular aperodic pulse
plot(t,f_t)
```

# Q4)Generate and display the following signal in four periods
$$x(t)=3cos(3\pi t+\frac{\pi}{3})$$

```julia
using LinearAlgebra
using Plots
A = 3;
B = 3*pi;
C = 0;
T = 2*pi/B;
delt = 0.01;
t = 0 : delt : 4*T;
ft = zeros(length(t));
ft = [A*cos(B*pi*t[i]+pi/3)+C for i in 1:length(t)];
plot(t,ft)
```
